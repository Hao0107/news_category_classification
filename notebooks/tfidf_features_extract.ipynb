**Thuật toán TF (Term Frequency - tần suất từ)**

$$TF(t,d)=\frac{f_{t,d}}{\sum_{t'}f_{t',d}}$$

**Trong đó**
- f_{t,d} là tần suất xuất hiện của từ t trong document d.
- sum_{t'}f_{t',d} là tổng số từ t' có trong document d.

**IDF (Inverse Document Frequency - Nghịch đảo tần suất tài liệu):**

$$IDF(t)=\log\left(\frac{1+N}{1+df_t}\right)+1$$

**Trong đó**
- N là tổng số bài báo.
- df_t là tổng số bài báo chứa từ t.
