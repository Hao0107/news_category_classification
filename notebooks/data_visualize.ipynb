Dựa trên kết quả trên ta có thể thấy:
- Số từ chỉ xuất hiện 1 lần (tên riêng, sai chính tả, mã số,...) là 17,322 từ (Chiếm 54.4%)
- Số từ chỉ xuất hiện trong <5 bài báo là 24,442 từ (Chiếm 76.8%)

Nhưng nếu ta loại bỏ các từ nhiễu, ít xuất hiện để giúp mô hình học tổng quát hơn không bị overfit thì ta sẽ loại bỏ toàn bộ các từ ít xuất hiện ấy ra khỏi từ điển(vocab) của dữ liệu.
Khi đó kích thước ma trận từ điển của ta có kích thước:
