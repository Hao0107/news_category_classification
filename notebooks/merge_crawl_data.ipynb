```python
import pandas as pd
from datetime import datetime
```

```python
data_path = "../data/crawl/raw_crawled_articles.csv"
new_data_path = f"../data/crawl/raw_crawled_articles_{datetime.now().strftime('%d%m%Y')}.csv"
```

```python
df = pd.read_csv(data_path)
new_df = pd.read_csv(new_data_path)
```

```python
df.shape, new_df.shape
```

```python
df_merged = pd.concat([df, new_df], ignore_index=True)
df_merged.shape
```

```python
df_merged.drop_duplicates(subset=['url'], inplace=True)
```

```python
df_merged.shape
```

```python
df_merged.to_csv("../data/crawl/raw_crawled_articles.csv", index=False)
```
